Import packages

```python

import csv
import requests
import pandas as pd
```

Setup webscraper via Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup

urls = [
    'https://www.pgatour.com/stats/detail/02674',
    'https://www.pgatour.com/stats/detail/02564',
    'https://www.pgatour.com/stats/detail/120',
    'https://www.pgatour.com/stats/detail/130',
    'https://www.pgatour.com/stats/detail/101',
    'https://www.pgatour.com/stats/detail/103',
    'https://www.pgatour.com/stats/detail/331',
    'https://www.pgatour.com/stats/detail/486',
    'https://www.pgatour.com/stats/detail/374',
    'https://www.pgatour.com/stats/detail/102',
]

header_mappings = [
    {
        'Avg': 'Avg_SG_T2G',
        'Measured Rounds': 'SG:Measured_Rounds'
        
    },
    {
        'Avg': 'Avg_SG_Putting',
        'Measured Rounds': 'Putting_Measured_Rounds'
        
    },
    {
        'Avg': 'Scoring_Avg',
        'Total Rounds': 'Total_Scoring_Rounds'
    },
    {
        '%': 'Scrambling_%'
    },
    {
        'Avg': 'Avg_Drive_Distance',
        'Total Distance': 'Total_Drive_Distance'
    },
    {
        '%': 'GIR_%',
        'Greens Hit': 'Total_Greens_Hit'
    },
    {
        'Avg': 'Avg_Prox_Hole_Fair',
        'Total Distance (feet)': 'Total_Distance_Fair',
        '# of Attempts': '#_of_Attempts_Fair'
    },
    {
        '%': 'Going_4_Green_%'
    },
    {
        'Avg DTP': 'Avg_DTP_ARG',
        'Total Distance (feet)': 'Total_Distance_ARG'
    },
    {
        '%': 'Driving_Acc_%'
    },
]

all_data = []
driver = webdriver.Chrome("C:\chromedriver")

# Iterate over the URLs and scrape the data from each page
for i, url in enumerate(urls):
    # Load the webpage
    driver.get(url)

    # Extract the page source after the JavaScript has executed
    page_source = driver.page_source

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(page_source, "lxml")

    div_ele = soup.find('div', class_='simplebar-content')
    table = div_ele.find('table')
    rows = table.find_all('tr')

    data = []

    for row in rows:
        row_data = [cell.text.strip() for cell in row.find_all("td")]
        data.append(row_data)

    heading = div_ele.find('thead')
    headers = heading.find_all('th')
    header_data = [cell.text.strip() for cell in headers]
    data.append(header_data)

    mapping = header_mappings[i]  # Mapping specific to the current URL
    for j in range(len(header_data)):
        if header_data[j] in mapping:
            header_data[j] = mapping[header_data[j]]

    all_data.append(data)

driver.quit()

# Print the headers and data for each URL
for i, data in enumerate(all_data):
    print(f"Headers for URL {urls[i]}:")
    print(data[-1])  # Print the last appended list which contains the headers

    print(f"Data for URL {urls[i]}:")
    for row in data[:-1]:  # Iterate over the data excluding the headers
        print(row)

    print('\n')

# Convert all_data into a DataFrame
dataframes = []
for data in all_data:
    headers = data[-1]  # Use the modified headers
    rows = data[:-1]
    df = pd.DataFrame(rows, columns=headers)
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=False)
combined_df.to_csv('New_data.csv', index=False)
```
