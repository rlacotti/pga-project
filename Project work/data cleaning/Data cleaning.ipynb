```python
import pandas as pd
```

Read in the data

```python
df = pd.read_csv('New_data.csv')
```

Remove unnecessary columns

```python

df.drop(['Putting_Measured_Rounds', 'Total Adjustment', 'Total_Scoring_Rounds', 
         'Total_Drive_Distance', 'Relative/Par', 'Total_Distance_Fair',
         'GIR Rank', 'RTP-Successes', 'RTP-Not going for the grn',
         'Total_Distance_ARG', 'Scrambling Rank'], axis=1, inplace=True)

df
```

Fill in the missing values

```python
grouped = df.groupby('Player')

# Fill missing values within each player group
filled_df = pd.DataFrame()

for player, group in grouped:
    filled_group = group.fillna(method='ffill').fillna(method='bfill')
    filled_df = pd.concat([filled_df, filled_group])

# Reset the index to convert the 'Player' column back to a regular column
filled_df = filled_df.reset_index(drop=True)

# Remove duplicate rows, keeping the last occurrence of each player's data
cleaned_df = filled_df.drop_duplicates(subset='Player', keep='last')
```

Show cleaned data

```python
cleaned_df
```

Remove more unnecessary columns

```python
cleaned_df.drop('Rank',  axis=1, inplace=True)
```

```python
cleaned_df.drop('Unnamed: 1', axis=1, inplace=True)
```

```python
cleaned_df
```

```python
cleaned_df.dropna(inplace=True)
```

Save the dataset to a new csv

```python
cleaned_df.to_csv('Cleaned_data.csv', index=False)
```
