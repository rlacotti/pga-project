Create the K means algorithm

```python
# Implement Algorithm
class KMeansClustering():
    def __init__(self, X, num_clusters, random_seed = None):
        self.K = num_clusters
        self.plot_figure_enabled = True
        self.max_iterations = 100
        self.num_examples, self.num_features = X.shape
        self.random_seed = random_seed

    def initialize_random_centroids(self, X):
        if self.random_seed is not None:
            np.random.seed(self.random_seed)
        centroids = np.zeros((self.K, self.num_features))

        for k in range(self.K):
            centroid = X[np.random.choice(range(self.num_examples))]
            centroids[k] = centroid

        return centroids
    

    def create_clusters(self, X, centroids):
        clusters = [[] for _ in range(self.K)]

        for point_idx, point in enumerate(X):
            closest_centroid = np.argmin(np.sqrt(np.sum( (point - centroids)**2, axis=1)))
            clusters[closest_centroid].append(point_idx)

        return clusters

    def calculate_new_centroids(self, clusters, X):
        centroids = np.zeros((self.K, self.num_features))

        for idx, cluster in enumerate(clusters):
            new_centroid = np.mean(X[cluster], axis=0)
            centroids[idx] = new_centroid

        return centroids

    def predict_cluster(self, clusters, X):
        y_pred = np.zeros(self.num_examples, dtype=int)

        for cluster_idx, cluster in enumerate(clusters):
            y_pred[cluster] = cluster_idx

        return y_pred.flatten()


    def plot_figure(self, X, y, player_names, centroids):
        plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.viridis)
        plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', s=100, color='red', label='Centroids')
            
        plt.xlabel('Avg_SG_T2G')
        plt.ylabel('Sg:OTT')
        plt.title('K-Means Clustering for PGA Tour Players')
        plt.colorbar(label = 'Cluster')
        plt.legend()
        plt.show(block=False)

    def classify_players(self, X, y_pred, player_names):
        
        long_bombers = [[] for _ in range(self.K)]
        iron_players = [[] for _ in range(self.K)]
        short_game_specialists = [[] for _ in range(self.K)]

        for idx, player in enumerate(X):
            cluster_label = int(y_pred[idx])
            if cluster_label == 0 and player[8] >= scaled_8 and player[1] >= scaled_1 and player[13] >= scaled_13:
                long_bombers[cluster_label].append(player_names[idx])
            elif cluster_label == 1 and player[10] <= scaled_10 and player[9] >= scaled_9 and player[2] >= scaled_2:
                iron_players[cluster_label].append(player_names[idx])
            elif cluster_label == 2 and player[12] <= scaled_12 and player[3] >= scaled_3 and player[6] >= scaled_6:
                short_game_specialists[cluster_label].append(player_names[idx])

                
        print("Long Bombers:")
        for cluster_label, players in enumerate(long_bombers):
            print(f"Cluster {cluster_label}: {players}")

        print("Iron Players:")
        for cluster_label, players in enumerate(iron_players):
            print(f"Cluster {cluster_label}: {players}")

        print("Short Game Specialists:")
        for cluster_label, players in enumerate(short_game_specialists):
            print(f"Cluster {cluster_label}: {players}")

    def cluster_scoring(self, X, y_pred, player_names):


        long_bombers_players = []
        long_bombers_scores = []

        iron_players_players = []
        iron_players_scores = []

        short_game_specialists_players = []
        short_game_specialists_scores = []

        for idx, player in enumerate(X):
            cluster_label = int(y_pred[idx])
            score = player[5]

            if cluster_label == 0:
                long_bombers_players.append(player_names[idx])
                long_bombers_scores.append(score)
            elif cluster_label == 1:
                iron_players_players.append(player_names[idx])
                iron_players_scores.append(score)
            elif cluster_label == 2:
                short_game_specialists_players.append(player_names[idx])
                short_game_specialists_scores.append(score)

        long_bombers_avg = np.mean(long_bombers_scores)
        iron_players_avg = np.mean(iron_players_scores)
        short_game_specialists_avg = np.mean(short_game_specialists_scores)

        print("Long Bombers Average Score:", long_bombers_avg)
        print("Iron Players Average Score:", iron_players_avg)
        print("Short Game Specialists Average Score:", short_game_specialists_avg)

    
        

    def fit(self, X, player_names):
        centroids = self.initialize_random_centroids(X)

        for it in range(self.max_iterations):
            clusters = self.create_clusters(X, centroids)

            previous_centroids = centroids
            centroids = self.calculate_new_centroids(clusters, X)

            diff = centroids - previous_centroids

            if not diff.any():
                print("Termination complete, K means converged successfully")
                break

        y_pred = self.predict_cluster(clusters, X)

    

        if self.plot_figure_enabled:
            self.plot_figure(X, y_pred, player_names, centroids)

        return y_pred
```

Test the significance of the findings

```python


import scipy.stats as stats

# Us the scores from each cluster
scores_long_bombers = data.loc[y_pred == 0, 'Scoring_Avg']
scores_iron_players = data.loc[y_pred == 1, 'Scoring_Avg']
scores_short_game_specialists = data.loc[y_pred == 2, 'Scoring_Avg']

# Perform ANOVA to test for difference among the means for multiple groups
anova_result = stats.f_oneway(scores_long_bombers, scores_iron_players, scores_short_game_specialists)

# print ANOVA results
print("ANOVA p-value:", anova_result.pvalue)

# If ANOVA suggests differences, perform pairwise t-tests
if anova_result.pvalue < 0.05:
    # Perform pairwise t-tests and correct for multiple comparisons 
    ttest_long_bombers_iron = stats.ttest_ind(scores_long_bombers, scores_iron_players)
    ttest_long_bombers_short_game = stats.ttest_ind(scores_long_bombers, scores_short_game_specialists)
    ttest_iron_short_game = stats.ttest_ind(scores_iron_players, scores_short_game_specialists)

    # Print t-test results
    print("T-Test Long Bombers vs. Iron Players p-value:", ttest_long_bombers_iron.pvalue)
    print("T-Test Long Bombers vs. Short Game Specialists p-value:", ttest_long_bombers_short_game.pvalue)
    print("T-Test Iron Players vs. Short Game Specialists p-value:", ttest_iron_short_game.pvalue)
```
